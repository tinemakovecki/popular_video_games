Data analysis on games by publisher:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
pd.options.display.max_rows = 20
```

```python
TwoK = pd.read_csv('2K Games.csv')
SEGA = pd.read_csv('SEGA.csv')
Activision = pd.read_csv('Activision.csv')
Activision[['Publisher']] = 'Activision'
EA = pd.read_csv('Electronic Arts.csv')
Beth = pd.read_csv('Bethesda Softworks.csv')
Klei = pd.read_csv('Klei Entertainment.csv')
Paradox = pd.read_csv('Paradox Interactive.csv')
Warner = pd.read_csv('Warner Bros.csv')
Ubisoft = pd.read_csv('Ubisoft.csv')
Valve = pd.read_csv('Valve.csv')

publishers = [SEGA, Activision, EA, Beth, Klei, Paradox, Warner, Ubisoft, Valve]
```

Combining data form different publishers into one table

```python
games = pd.concat(publishers)
games.index = [i for i in range(len(games))]
```

Changing column names for easier syntax:

```python
games = games.rename(columns={'Release year': 'Year', 'Average playtime': 'Average_playtime', 'Median playtime': 'Median_playtime'})
games
```

Average game rating for each publisher

```python
By_publisher = games.groupby('Publisher')
By_publisher['Userscore'].mean().plot(kind = 'bar', title = 'developer game quality')
plt.show()
```

```python
year = games[games.Year == 2016]
year.sort_values('Metascore', ascending = False)[0:5]
```

Example of a game companies average game quality through time, Activision is used for the example

```python
activision = games[games.Publisher == 'Activision']
qual = activision.groupby('Year').Userscore.mean()
qual.plot(title = 'Activision game ratings through time')
plt.show()
```

We can use the data for median and average playtime to find the most "polarizing" games, which are games where most owners have very little playtime, but some players have a large amount of playtime.

```python
game_pol = games[['Game', 'Price', 'Userscore', 'Metascore', 'Owners', 'Average_playtime', 'Median_playtime']]
game_pol['Ratio'] = game_pol.Average_playtime / game_pol.Median_playtime
game_pol
```

```python
game_pol.sort_values('Ratio', ascending=False)
```

We can notice that a lot of the games with very large differences in playtime are free. Free games can have a lot of "owners" which downloaded the game only to try it for a short time. This lowers the median playtime and shows in this data.

Now we're adding the data for game genres as well.

```python
action = pd.read_csv('Action.csv')
simulation = pd.read_csv('Simulation.csv')
sports = pd.read_csv('Sports.csv')
MMO = pd.read_csv('MMO.csv')
RPG = pd.read_csv('RPG.csv')
strategy = pd.read_csv('Strategy.csv')
adventure = pd.read_csv('Adventure.csv')
indie = pd.read_csv('Indie.csv')
```

```python
Genres = [action, simulation, sports, MMO, RPG, strategy, adventure, indie]
by_genres = pd.concat(Genres)
```

```python
by_genres = by_genres.rename(columns={'Release year': 'Year', 'Average playtime': 'Average_playtime', 'Median playtime': 'Median_playtime'})
```

```python
by_genres
```

We're not really interested in games without any kind of rating, since those are either very unpopular or beta tests, etc. So we're going to remove those.

```python
new_genres = by_genres[(by_genres.Userscore != '/') & (by_genres.Userscore != '')]
```

Below we can see that the largest number of games is in the inde genre, followed by action and adventure.

```python
new_genres.groupby('Genre').Game.count().plot(kind='pie', title='Number of games by genre')
plt.show()
```

```python
new_genres[['Userscore']] = new_genres[['Userscore']].apply(pd.to_numeric)
```

```python
new_genres.groupby('Genre').Userscore.mean().plot(kind='bar', title='Average rating by genre')
plt.show()
```

The largest deviation in rating is MMO games, which on average have the lowest rating,the other genres don't differ to much. Simulation games get second worst ratings and indie games have best ratings on average.

As an example, we're going to find out, which genre of games Bethesda Softworks publishes most often

```python
beth_games = [game for game in Beth.Game]
```

```python
beth_genre_count = {'action':0, 'adventure':0, 'indie':0, 'MMO':0, 'simulation':0, 'RPG':0, 'sports':0, 'strategy':0}
for game in beth_games:
    matches = [genre for genre in by_genres[by_genres.Game == game].Genre]
    for x in matches:
        beth_genre_count[x] = beth_genre_count[x] + 1
```

```python
pd.Series(beth_genre_count).plot(kind='pie', title='Bethesda games by genre')
plt.show()
```

In the graph we can see that almost all Bethesda games are either action or RPG.
